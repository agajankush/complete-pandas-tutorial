#### String Operations

```python
# Copy the bios data into a new DataFrame so we dont damage the original data
bios_new = bios.copy()
```

```python
bios_new.head()
```

```python
# Get just the first name and create a new column
bios_new["first_name"] = bios_new["name"].str.split(" ").str[0]
```

```python
bios_new.info()
```

```python
# We can see that the date field is not a datetime object and if we convert it into a datetime object 
# It will be easier to use inbuilt pandas functions
bios_new["born_date"] = pd.to_datetime(bios_new["born_date"])
```

```python
bios_new["born_year"] = bios_new["born_date"].dt.year
```

```python
bios_new.head()
```

```python
bios_new.loc[(bios_new["first_name"] == "Keith"), ["first_name", "born_year"]]
```

```python
bios_new.to_csv("./data/bios_new.csv", index=False)
```
